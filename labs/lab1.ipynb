
Explicação de cada um dos headers de REQUEST:

- `Host`: É o host do servidor e a porta. Se a porta não estiver inclusa, usa a porta 443 no caso de https e 80 no caso de http
- `User-agent`:É uma string que permite aos servidores e a rede identificarem a aplicação, sistema operacional, versão do User Agent. No caso, do navegador, o user Agent é o browser.  
- `Accept`: Específica o tipo de conteúdo que o cliente consegue entender.
- `Accept-Language`: Indica a linguagem padrão que o usuário prefere.
- `Accept-Encoding`: Indica o encoding do conteúdo, geralmente é especificado um algoritmo de compressão
- `Referer`: É um endereço absoluto ou parcial da página que fez o request. Isso permite o servidor a página que os usuários estão visitando para conseguir esse conteúdo. Essa informação pode ser usada para análise posteriormente.
- `Connection`: Controla se a conexão permanece aberta após o request.
- `Cookie`: Contém o Cookie associado ao servidor.
- `Sec-Fetch-Dest`: Indica o destino do request. Por exemplo, se o destino for do tipo `audio` então só pode ser pedido dados do tipo `audio`.
- `Sec-Fetch-Mode`: Permite o servidor diferenciar entre requests vindos da navegação de usuário entre páginas HTML e requests para carregar imagens e outros recursos. No caso do request acima, ele é `no-cors`, então os headers precisam ser simples e o método só pode ser GET, HEAD ou POST
- `Sec-Fetch-Site`: Indica a relação entre a origem do iniciador do request e a origem do recurso requisitado. Esse header diz para o servidor se o request para um recurso vem da mesma origem ou não.
- `TE`: Especifica a codificação de transferência que o agente de usuário está propício a aceitar.

```
HTTP/2 200 OK
accept-ranges: bytes
cross-origin-resource-policy: cross-origin
cross-origin-opener-policy-report-only: same-origin; report-to="static-on-bigtable"
report-to: {"group":"static-on-bigtable","max_age":2592000,"endpoints":[{"url":"https://csp.withgoogle.com/csp/report-to/static-on-bigtable"}]}
content-length: 44570
x-content-type-options: nosniff
server: sffe
x-xss-protection: 0
date: Tue, 04 Oct 2022 13:41:05 GMT
expires: Wed, 04 Oct 2023 13:41:05 GMT
cache-control: public, max-age=31536000
last-modified: Tue, 23 Aug 2022 17:46:27 GMT
content-type: image/webp
age: 110938
alt-svc: h3=":443"; ma=2592000,h3-29=":443"; ma=2592000,h3-Q050=":443"; ma=2592000,h3-Q046=":443"; ma=2592000,h3-Q043=":443"; ma=2592000,quic=":443"; ma=2592000; v="46,43"
X-Firefox-Spdy: h2    
 ```



Explicação de cada um dos headers de RESPONSE:

- `accept-ranges`: É um marcador usado pelo servidor para indicar que ele suporta requisições parciais. O valor deste campo indica a unidade utilizada para definir o tamanho. Isso pode ajudar o navegador a retomar um download interrompido no lugar de iniciar um do zero novamente.
- `cross-origin-resource-policy`: Indica um desejo que o navegador bloqueie requisições que não sejam entre origens e/ou entre sites para um dado recurso. Por exemplo, `same-origin` significa que apenas requests da mesma origem podem ler o resource.
- `cross-origin-opener-policy-report-only`: Garante que requisições entre origens sejam bloqueadas para um dado recurso.
- `x-content-type-options`: Indica que os MME Types no header `content-type` não podem ser alterados.
- `server`: Informações do software usado pelo servidor.
- `x-xss-protection`: É uma funcionalidade do Internet Explorer, Chrome e Safari que impede páginas de carregarem quando eles detectam ataques de scripting entre sites.
- `date`: Data que a mensagem foi gerada.
- `expires`: Data que a resposta é considerada velha.
- `cache-control`: O campo de cabeçalho genérico Cache-Control é usado para especificar diretivas para mecanismos de cache tanto em requisições quanto em respostas.
- `last-modified`: Data que o servidor modificou o recurso pela última vez.
- `content-type`: Indica o tipo de arquivo do recurso.
- `age`: Tempo para que o objeto seja removido do cache de um proxy.
- `alt-svc`: Indica que serviços alternativos que podem ser utilizados para alcançar o mesmo recurso.
- `X-Firefox-Spdy`: É um header adicionado pelo firefox para identificar um request de http2.

<span style="color: blue">

2. Faça  uma  pesquisa  no  Google  utilizando  um  termo  qualquer,  porém  acrescente  na  caixa  de  busca  a  seguinte  informação: filetype:pdf. Assim sua caixa de busca conterá: “termo” filetype:pdf. A partir do primeiro link devolvido, obtenha seu conteúdo e inspecione o resultado. Qual o valor para o header “content-type”? O que este valor mesmo significa? Inclua em seu relatório tanto o request, quanto o response obtido nesta operação.
</span>


Como pode ver abaixo, o `content-type` vem como `appalication/pdf`. Esse header especifica o tipo padrão de resposta como sendo PDF.

<span style="color: blue">
3. Repita   o   mesmo   procedimento   acima,   porém   adicione   na   caixa   de   busca   a   seguinte   informação:   filetype:xls. Assim sua caixa de busca conterá: “termo” filetype:xlsA partir do primeiro link devolvido, obtenha seu conteúdo e inspecione o resultado. Qual o valor para o header “content-type”? O que este valor mesmo significa? Inclua em seu relatório tanto o request, quanto o responseobtido nesta operação
</span>


Abaixo, o conteúdo. esse `content-type: application/vnd.ms-excel` específica que o conteúdo da resposta será uma planilha de excel, mais especificamente uma planilha em `.xls` ou `.xlsx`

<span style="color: blue">
4. Digite no navegador o seguinte endereço: www.ufu.br/flavio. Veja o que acontece e comente o resultado após inspecionar  os  dados  no  “Web  Developer  Tools”.  Inclua  em  seu  relatório  tanto  o  request,  quanto  o  responseobtido nesta operação
</span>


A paǵina não foi encontrada, acabou dando um not found 404. Nesse caso, o REQUEST vai vazio, mas o response vem com o html da página de Not Found padrão.

<span style="color: blue">
5. Inspecione   o   request   e   o   response   para   os   seguintes   endereços:   http://www.portal.facom.ufu.br/   e   http://www.portal.facom.ufu.br:80  .  Existe  alguma  diferença  entre  request  e  response  em  ambos  os  casos?  Explique este comportamento. Inclua em seu relatório tanto o request, quanto o response obtido nesta operação. 
</span>


Como é possível ver abaixo, ao especificar a porta 80, não consegui acessasr o site, ou melhor, o RESPONSE veio vazio. Isso se deve ao fato de quem em websites geralmente a porta 443 é a que fica aberta por padrão. Se o servidor não estiver servindo mais nada além do site, o firewall desabilita a conexão com as outras restantes.

<span style="color: blue">
6. Inspecione  o  resultado  de  um  pedido  no  seguinte  endereço:  http://www.facom.ufu.br:443/.  Explique  o  que  ocorreu? Inclua em seu relatório tanto o request, quanto o response obtido nesta operação. 
</span>   
 

Como é possível observar abaixo, nesse caso funciona, porque a porta padrão aberta é a 443.

<span style="color: blue">

7. Acesse seguinte endereço: http://www.portal.facom.ufu.br:8080/. O que ocorreu? Explique o comportamento do “Web  Developer  Tools”  nesta  situação.  Qual  alteração  seria  necessária  no  servidor  a  fim  de  que  o  endereço  devolvesse algum resultado? 

</span>   


Ao acessar a porta 8080, a página fica carregando indefinidamente sem devolver nada. O servidor deveria habilitar essa porta para que tivesse alguma resposta.

<span style="color: blue">

8. Inspecione o que ocorre com os seguintes endereços: http://pardal.dr.ufu.br/ e http://pardal.dr.ufu.br:8080/. O que  ocorreu  no  primeiro  endereço?  E  no  segundo?  Explique  a  diferença  de  comportamento.  O  que  significa  o  número  8080  adicionado  no  endereço?  Inclua  em  seu  relatório  tanto  o  request,  quanto  o  response  obtido  em  cada uma destas operações. 
    
</span>
