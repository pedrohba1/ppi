
Explicação de cada um dos headers de REQUEST:

- `Host`: É o host do servidor e a porta. Se a porta não estiver inclusa, usa a porta 443 no caso de https e 80 no caso de http
- `User-agent`:É uma string que permite aos servidores e a rede identificarem a aplicação, sistema operacional, versão do User Agent. No caso, do navegador, o user Agent é o browser.  
- `Accept`: Específica o tipo de conteúdo que o cliente consegue entender.
- `Accept-Language`: Indica a linguagem padrão que o usuário prefere.
- `Accept-Encoding`: Indica o encoding do conteúdo, geralmente é especificado um algoritmo de compressão
- `Referer`: É um endereço absoluto ou parcial da página que fez o request. Isso permite o servidor a página que os usuários estão visitando para conseguir esse conteúdo. Essa informação pode ser usada para análise posteriormente.
- `Connection`: Controla se a conexão permanece aberta após o request.
- `Cookie`: Contém o Cookie associado ao servidor.
- `Sec-Fetch-Dest`: Indica o destino do request. Por exemplo, se o destino for do tipo `audio` então só pode ser pedido dados do tipo `audio`.
- `Sec-Fetch-Mode`: Permite o servidor diferenciar entre requests vindos da navegação de usuário entre páginas HTML e requests para carregar imagens e outros recursos. No caso do request acima, ele é `no-cors`, então os headers precisam ser simples e o método só pode ser GET, HEAD ou POST
- `Sec-Fetch-Site`: Indica a relação entre a origem do iniciador do request e a origem do recurso requisitado. Esse header diz para o servidor se o request para um recurso vem da mesma origem ou não.
- `TE`: Especifica a codificação de transferência que o agente de usuário está propício a aceitar.
